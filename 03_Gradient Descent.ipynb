#### learning?
- find 'w' that minimize the loss





```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0 # random value

def forward(x):
    return w * x

def loss(x, y):
    y_pred = forward(x)
    
    return (y_pred-y) ** 2 # MSE(Mean Square Error)로 가정

def gradient(x, y): # d_loss / d_w
    return 2 * x * (w*x - y)
```

```python
print(f"predict (before training) x = {4}, y_pred = {forward(4)}")
```

```python
lr = 0.01 # learning rate(α : 알파)

for epoch in range(100): # epoch = 100    
    for x_val, y_val in zip(x_data, y_data):
        grad = gradient(x_val, y_val)
        w = w - lr * grad
        print("\t grad : ", x_val, y_val, grad)
        l = loss(x_val, y_val)
        
    print(f"epoch : {epoch}, w : {w}, loss : {l}")

print(f"predict (after training) x = {4}, y_pred = {forward(4)}")
```
